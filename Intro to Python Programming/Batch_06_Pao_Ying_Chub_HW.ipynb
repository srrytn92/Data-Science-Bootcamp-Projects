Import Library

```python
import random as rd
```

Set Variables

```python
bot = ["Rock", "Paper", "Scissors"]
win = 0
draw = 0
loss = 0
round = 0
```

Game

```python
print(" Hi! Welcome to Rock Paper Scissors")
print("""
Do you want to play games?
press 1 = yes 
press any key = no
""")
start = int(input("Please select : "))
while start == 1 :
   print ("""
   Please select 
   press 1 = Rock  
   press 2 = Paper 
   press 3 = Scissors 
   press any Key to Exit""")
   user = int(input("Please select : "))
   
   ## User choose 1 Rock
   if user == 1 :
     bot_s = rd.choice(bot)
     if bot_s == "Rock" :
       print(f"""
       You choose : Rock 
       Bot choose : {bot_s}
       You Draw""")
       round += 1
       draw += 1
     elif bot_s == "Paper" :
       print(f"""
       You choose : Rock 
       Bot choose : {bot_s}
       You Loss""")
       round += 1
       loss += 1
     elif bot_s == "Scissors" :
       print(f"""
       You choose : Rock 
       Bot choose : {bot_s}
       You Win""")
       round += 1
       win += 1

   ## User choose 2 Paper
   elif user == 2 :
     bot_s = rd.choice(bot)
     if bot_s == "Rock" :
       print(f"""
       You choose : Paper 
       Bot choose : {bot_s}
       You Win""")
       round += 1
       win += 1
     elif bot_s == "Paper" :
       print(f"""
       You choose : Paper 
       Bot choose : {bot_s}
       You Draw""")
       round += 1
       draw += 1
     elif bot_s == "Scissors" :
       print(f"""
       You choose : Paper 
       Bot choose : {bot_s}
       You Loss""")
       round += 1
       loss += 1

   ## User choose 3 Scissors
   elif user == 3 :
     bot_s = rd.choice(bot)
     if bot_s == "Rock" :
       print(f"""
       You choose : Scissors 
       Bot choose : {bot_s}
       You Loss""")
       round += 1
       loss += 1
     elif bot_s == "Paper" :
       print(f"""
       You choose : Scissors 
       Bot choose : {bot_s}
       You Win""")
       round += 1
       win += 1
     elif bot_s == "Scissors" :
       print(f"""
       You choose : Scissors 
       Bot choose : {bot_s}
       You Draw""")
       round += 1
       draw += 1 

   else :
    print(f"""
    result
    round : {round} time
    win : {win} time
    draw : {draw} time
    loss : {loss} time
    """)
    break
print("See you again next time")
```
